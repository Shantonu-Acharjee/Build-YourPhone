```python
import requests
from bs4 import BeautifulSoup





class MobilePhone:

    def __init__(self) :
        self.brawser_header = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.110 Safari/537.36"}
        self.url = 'https://www.gsmarena.com.bd/'

        # Varibles list
        self.brands_name_list_ = []

        self.this_brands_is_not_found_ = []



        # Functions
        self.__brand_name_list()
        self.__get_all_phone_by_brand()





    def __brand_name_list(self):
        url = self.url + 'brands/'
        webPage = requests.get(url, headers= self.brawser_header).text
        soup = BeautifulSoup(webPage, 'lxml')
        cards = soup.find_all('div', class_ = 'product-thumb')

        for i in cards:

            if i.find_all('a')[1].text.strip().lower().split()[0] in ['general', 'geo', 'google', 'nothing']:
                continue

            self.brands_name_list_.append(i.find_all('a')[1].text.strip().lower().split()[0])

        
        for i in ["general-mobile", "geo-phone", "google-phone", "nothing-phone"]:
            self.brands_name_list_.append(i)


    

    def __get_all_phone_by_brand(self):
        for i in self.brands_name_list_:
            url = self.url + i
            webPage = requests.get(url, headers= self.brawser_header).text
            soup = BeautifulSoup(webPage, 'lxml')


            if soup.find('h1', class_ = 'title').text.strip().lower().split()[0] == i:
                print(i,': is ok')


            else:
                self.this_brands_is_not_found_.append(i)
                print(i, ': this brand is not found')
        
```

```python
a = MobilePhone()
```
