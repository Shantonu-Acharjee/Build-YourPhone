```python
import requests
from bs4 import BeautifulSoup
import pandas as pd





class MobilePhone:

    def __init__(self) :
        self.brawser_header = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.110 Safari/537.36"}
        self.url = 'https://www.gsmarena.com.bd/'

        # Varibles list
        self.brands_name_list_ = []

        self.this_brands_is_not_found_ = []
        self.incomplate_brands_name_ = ['general', 'geo', 'google', 'nothing']
        self.complate_brands_name_ = ["general-mobile", "geo-phone", "google-phone", "nothing-phone"]
        self.one_brend_all_phone_list_ = []



        # Functions
        self.__brand_name_list()
        self.__get_all_phone_by_brand()
        self.__single_phone_detals_valadity_cheek()





    def __brand_name_list(self):
        url = self.url + 'brands/'
        webPage = requests.get(url, headers= self.brawser_header).text
        soup = BeautifulSoup(webPage, 'lxml')
        cards = soup.find_all('div', class_ = 'product-thumb')

        for i in cards:

            if i.find_all('a')[1].text.strip().lower().split()[0] in self.incomplate_brands_name_:
                continue

            self.brands_name_list_.append(i.find_all('a')[1].text.strip().lower().split()[0])

        
        for i in self.complate_brands_name_:
            self.brands_name_list_.append(i)









    

    def __get_all_phone_by_brand(self):

        
        for i in self.brands_name_list_:
            url = self.url + i
            webPage = requests.get(url, headers= self.brawser_header).text
            soup = BeautifulSoup(webPage, 'lxml')


            if soup.find('h1', class_ = 'title').text.strip().lower().split()[0] == i or soup.find('h1', class_ = 'title').text.strip().lower().split()[0] in self.incomplate_brands_name_:
                print(i,': is ok')

                e = 0
                while e < 6:
                    url = self.url + str(i) + '/' + str(e+1)
                    webPage = requests.get(url, headers= self.brawser_header).text
                    soup = BeautifulSoup(webPage, 'lxml')
                    print(url)



                    try:

                        if soup.find_all('div', class_ = 'mobile_name')[0].text.strip().lower().replace('+', '').replace(' ','-').replace('(', '').replace(')', '').replace('.', '') in self.one_brend_all_phone_list_:
                            print(e, 'Do not find')
                            e = 6
                            continue


                        one_brend_all_phone_list = soup.find_all('div', class_ = 'mobile_name')

                        for j in range(len(one_brend_all_phone_list)):

                            self.one_brend_all_phone_list_.append(one_brend_all_phone_list[j].text.strip().lower().replace('+', '').replace(' ','-').replace('(', '').replace(')', '').replace('.', ''))


                        e = e+1

                    except:
                        e = 7
                        print(i,'Mobile not avable abable for this barand')

    

            

            else:
                self.this_brands_is_not_found_.append(i)
                print(i, ': this brand is not found')














    def __single_phone_detals_valadity_cheek(self):
        

        for i in self.one_brend_all_phone_list_:
            url = self.url + str(i)
            webPage = requests.get(url, headers= self.brawser_header).text
            soup = BeautifulSoup(webPage, 'lxml')

            try:
                if soup.find('h1', class_ = 'ptitle').text.strip().lower().replace('+', '').replace(' ','-').replace('(', '').replace(')', '').replace('.', '') == i:
                    print(i,': Mobile found')

            except:       
                print(i, 'Mobile note found x--x--x')
                
                del self.one_brend_all_phone_list_[self.one_brend_all_phone_list_.index(i)]


        

        df = pd.DataFrame(self.one_brend_all_phone_list_, columns= ['Mobile_Model'])
        df.to_csv('mobilePhone.csv', index= False)
        
```
