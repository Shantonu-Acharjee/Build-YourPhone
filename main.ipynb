```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
brawser_header = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.110 Safari/537.36"}
url = 'https://www.gsmarena.com.bd/brands'

webPage = requests.get(url, headers= brawser_header).text
soup = BeautifulSoup(webPage, 'lxml')
```

```python
soup.find_all('div', class_ = 'text-center')
```

```python
rows = soup.find_all('div', class_ = 'product-thumb')
```

```python
len(rows)
```

```python
print(rows)
```

```python
brands_name = []
for i in rows:
    brands_name.append(i.find_all('a')[1].text.strip().lower())
```

```python
brands_name
```

```python
len(brands_name)
```

```python
df = pd.DataFrame(brands_name, columns= ['brands_name'])
df.to_csv('mobilePhone.csv', index= False)
```
