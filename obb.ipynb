```python
import requests
from bs4 import BeautifulSoup
#import pandas as pd


class YourPhone:
    def __init__(self):
        
        self.brawser_header = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.110 Safari/537.36"}
        self.url = 'https://www.gsmarena.com.bd/'

        self.brands_name_list = []
        self.mobiles_phone_list = []
        self.single_phone_link_list = []


        self.phone_photo_link = []
        self.phone_price = []
        self.network_type = []
        self.launch_date = []


        self.brand_name()
        self.brand_phone()
        self.singleMobile()
        print('Done')







    def brand_name(self):
        url = self.url + 'brands/'
        webPage = requests.get(url, headers= self.brawser_header).text
        soup = BeautifulSoup(webPage, 'lxml')
        rows = soup.find_all('div', class_ = 'product-thumb')

        for i in rows:
            self.brands_name_list.append(i.find_all('a')[1].text.strip().lower().split()[0])






    def brand_phone(self):
        for i in self.brands_name_list:
            url = self.url + i
            webPage = requests.get(url, headers= self.brawser_header).text
            soup = BeautifulSoup(webPage, 'lxml')
            one_brend_all_phone_list = soup.find_all('div', class_ = 'mobile_name')

            for i in range(len(one_brend_all_phone_list)):

                self.mobiles_phone_list.append(one_brend_all_phone_list[i].text.strip().lower().replace('+', '').replace(' ','-').replace('(', '').replace(')', '').replace('.', ''))

            #print(self.mobiles_phone_list)
            #break -----------------------------------------


    

    def singleMobile(self):

        for i in self.mobiles_phone_list:
            url = self.url + i
            self.single_phone_link_list.append(url)
            webPage = requests.get(url, headers= self.brawser_header).text
            soup = BeautifulSoup(webPage, 'lxml')

            
            if soup.find('h1', class_ = 'ptitle').text.strip().lower().replace('+', '').replace(' ','-').replace('(', '').replace(')', '').replace('.', '') == str(i):

                


                

                # Simge phone photo link
                if soup.find_all('div', class_ = 'area')[0].find('img')['src']:
                    self.phone_photo_link.append(soup.find_all('div', class_ = 'area')[0].find('img')['src'])

                else:
                    self.phone_photo_link.append('NaN')






                # Single phone price
                print(i)
                if soup.find_all('td', class_ = 'specs_name2')[3].text.strip().index('.'):
                    
        
                    a = soup.find_all('td', class_ = 'specs_name2')[3].text.strip().index('.')
                    b = int(soup.find_all('td', class_ = 'specs_name2')[3].text.strip()[0:a].replace(',', ''))

                    if b <= 0 :
                        self.phone_price.append('NaN')

                    else:
                        self.phone_price.append(b)


                else:
                    self.phone_price.append('NaN')





                network_state = 0
                launch_date_state = 0

                #--------------------------------------------------------
                for i in range(len(soup.find_all('td', class_ = 'specs_name2'))):


                    try:

                        #network type
                        if soup.find_all('div', class_ = 'specs_name')[i].text.strip() == 'Network Type':
                            self.network_type.append(soup.find_all('td', class_ = 'specs_name2')[i+6].text.strip().replace('/', '-').replace(' ', ''))
                            network_state = 1




                        # Launch date
                        if soup.find_all('div', class_ = 'specs_name')[i].text.strip() == 'Launch Announcement':

                            if soup.find_all('td', class_ = 'specs_name2')[i+6].text.strip().replace(',', '-').replace(' ', '') == 'Notannouncedyet':
                                self.launch_date.append('NaN')
                                launch_date_state = 1
                                continue

                            self.launch_date.append(soup.find_all('td', class_ = 'specs_name2')[i+6].text.strip().replace(',', '-').replace(' ', ''))
                            launch_date_state = 1
                            #print(soup.find_all('td', class_ = 'specs_name2')[i+6].text.strip().replace(',', '-').replace(' ', ''))












                        

                    except:
                        pass






                if network_state == 0:
                    self.network_type.append('NaN')

                if launch_date_state == 0:
                    self.launch_date.append('NaN')




                #break

            else:
                del self.mobiles_phone_list[self.mobiles_phone_list.index(i)]


            

        

    
        
        
```
